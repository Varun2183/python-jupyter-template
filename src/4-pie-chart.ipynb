### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

def validate_pie_data(labels, sizes):
    if not labels or not sizes:
        raise ValueError("Labels and sizes cannot be empty.")
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must be of equal length.")
    if not all(isinstance(i, (int, float)) and i > 0 and i is not None for i in sizes):
        raise ValueError("Sizes must be positive numbers and non-null.")

try:
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    validate_pie_data(labels, sizes)

    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, wedgeprops={'edgecolor': 'black'})
    plt.title("Pie Chart")
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
