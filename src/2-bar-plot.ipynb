### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

def validate_bar_data(categories, values):
    if not categories or not values:
        raise ValueError("Categories and values cannot be empty.")
    if len(categories) != len(values):
        raise ValueError("Categories and values must have the same length.")
    if not all(isinstance(v, (int, float)) and v is not None for v in values):
        raise ValueError("Values must be numeric and non-null.")

try:
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    validate_bar_data(categories, values)

    plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
    plt.title("Bar Chart")
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
