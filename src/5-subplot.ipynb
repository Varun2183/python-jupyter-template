### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Validation functions
def validate_data(x, y):
    if not x or not y:
        raise ValueError("X and Y data cannot be empty.")
    if len(x) != len(y):
        raise ValueError("X and Y must have the same length.")
    if not all(isinstance(i, (int, float)) for i in x + y):
        raise TypeError("X and Y values must be numeric.")

def validate_bar_data(categories, values):
    if not categories or not values:
        raise ValueError("Categories and values cannot be empty.")
    if len(categories) != len(values):
        raise ValueError("Categories and values must have the same length.")
    if not all(isinstance(i, (int, float)) for i in values):
        raise TypeError("Values in bar chart must be numeric.")

def validate_scatter_data(x, y):
    validate_data(x, y)  # Reuse line plot validation

def validate_pie_data(labels, sizes):
    if not labels or not sizes:
        raise ValueError("Labels and sizes cannot be empty.")
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must have the same length.")
    if not all(isinstance(i, (int, float)) and i > 0 for i in sizes):
        raise ValueError("Sizes must be positive numeric values.")

# Data for each plot
x = list(range(11))
y = [i**2 for i in x]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Validate data before plotting
try:
    validate_data(x, y)
    validate_bar_data(categories, values)
    validate_scatter_data(x_scatter, y_scatter)
    validate_pie_data(labels, sizes)
    
    # Creating a 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))

    # Line Plot
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')
    axes[0, 0].set_title('Line Plot')
    axes[0, 0].set_xlabel('X-axis')
    axes[0, 0].set_ylabel('Y-axis')
    axes[0, 0].grid(True)
    axes[0, 0].legend()

    # Bar Plot
    colors = ['red', 'blue', 'green', 'purple']
    axes[0, 1].bar(categories, values, color=colors)
    axes[0, 1].set_title('Bar Plot')
    axes[0, 1].set_xlabel('Categories')
    axes[0, 1].set_ylabel('Values')

    # Scatter Plot
    axes[1, 0].scatter(x_scatter, y_scatter, color='magenta', label='Data Points')
    for i, txt in enumerate(zip(x_scatter, y_scatter)):
        axes[1, 0].annotate(txt, (x_scatter[i], y_scatter[i]))
    axes[1, 0].set_title('Scatter Plot')
    axes[1, 0].set_xlabel('X-axis')
    axes[1, 0].set_ylabel('Y-axis')
    axes[1, 0].legend()

    # Pie Chart
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'pink'])
    axes[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")
except TypeError as te:
    print(f"TypeError: {te}")
```
