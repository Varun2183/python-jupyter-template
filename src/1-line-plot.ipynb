### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11, 1)

# Compute y values as the square of x
y = x ** 2

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='$y = x^2$')

# Customize the plot
plt.title('Square of x', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values ($x^2$)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```
