### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(x, y):
    if not x or not y:
        raise ValueError("X and Y values cannot be empty.")
    if len(x) != len(y):
        raise ValueError("X and Y must have the same length.")
    if not all(isinstance(i, (int, float)) and i is not None for i in x + y):
        raise ValueError("X and Y values must be numeric and non-null.")

try:
    x = np.arange(0, 11, 1)
    y = x**2
    validate_data(x.tolist(), y.tolist())

    plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = xÂ²')
    plt.title("Line Plot")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.legend()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
