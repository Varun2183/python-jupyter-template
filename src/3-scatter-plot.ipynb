### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

def validate_scatter_data(x, y):
    if not x or not y:
        raise ValueError("X and Y values cannot be empty.")
    if len(x) != len(y):
        raise ValueError("X and Y must have the same length.")
    if not all(isinstance(i, (int, float)) and i is not None for i in x + y):
        raise ValueError("X and Y values must be numeric and non-null.")

try:
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]
    validate_scatter_data(x, y)

    plt.scatter(x, y, color='blue', marker='o', label='Data points')
    for i in range(len(x)):
        plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')

    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.grid(True)
    plt.legend()
    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
